```python
import numpy as np
```

```python
cat_matrix = np.load('../../data/extracts/categories_indexed.npy')
```

```python
cat_matrix.shape
```

```python
def get_nextcat(cm,cat_input,cat_lvl,cat_lvl_max):
    set_base_mapping = set()
    for j in range(cm.shape[0]):
        set_base_mapping = set_base_mapping | set(cat_matrix[j][:,0])
    return set_base_mapping
```

```python
dict_dir = {}
# Create base mapping (1st lvl)

set_base_mapping = set()

for j in range(cat_matrix.shape[0]):
    set_base_mapping = set_base_mapping | set(cat_matrix[j][:,0])
    set_base_mapping = set_base_mapping - set([0])

print set_base_mapping

for cat in set_base_mapping:
    dict_dir[cat] = {}
    
print dict_dir.keys()
```

```python
# create (2nd lvl)
for cat in dict_dir.keys():
    set_mapping = set()

    for j in range(cat_matrix.shape[0]):
        set_mapping = set_mapping | set(cat_matrix[j][np.where((cat_matrix[j][:,:1] == (cat)).all(axis=1))][:,1])
        set_mapping = set_mapping - set([0])
    
    for setcat in set_mapping:
        dict_dir[cat][setcat] = {}
    print cat
    print set_mapping
```

```python
# create (3nd lvl)
for cat0 in dict_dir.keys():
    for cat in dict_dir[cat0].keys():
        set_mapping = set()

        for j in range(cat_matrix.shape[0]):
            set_mapping = set_mapping | set(cat_matrix[j][np.where((cat_matrix[j][:,:2] == (cat0,cat)).all(axis=1))][:,2])
            set_mapping = set_mapping - set([0])
            
        for setcat in set_mapping:
            dict_dir[cat0][cat][setcat] = {}
        print cat0,cat
        print set_mapping
```

```python
# create (4nd lvl)
for cat0 in dict_dir.keys():
    for cat1 in dict_dir[cat0].keys():
        for cat in dict_dir[cat0][cat1].keys():
            set_mapping = set()

            for j in range(cat_matrix.shape[0]):
                set_mapping = set_mapping | set(cat_matrix[j][np.where((cat_matrix[j][:,:3] == (cat0,cat1,cat)).all(axis=1))][:,3])
                set_mapping = set_mapping - set([0])
                
            for setcat in set_mapping:
                dict_dir[cat0][cat1][cat][setcat] = {}
            print cat0,cat1,cat
            print set_mapping
```

```python
# create (5nd lvl)
for cat0 in dict_dir.keys():
    for cat1 in dict_dir[cat0].keys():
        for cat2 in dict_dir[cat0][cat1].keys():
            for cat in dict_dir[cat0][cat1][cat2].keys():
                set_mapping = set()

                for j in range(cat_matrix.shape[0]):
                    set_mapping = set_mapping | set(cat_matrix[j][np.where((cat_matrix[j][:,:4] == (cat0,cat1,cat2,cat)).all(axis=1))][:,4])
                    set_mapping = set_mapping - set([0])

                for setcat in set_mapping:
                    dict_dir[cat0][cat1][cat2][cat][setcat] = {}
                print cat0,cat1,cat
                print set_mapping
```

```python
np.save('../../data/derived/dict_dir.npy',dict_dir)
```
