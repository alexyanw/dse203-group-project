```python
# connect
import psycopg2
import json
import numpy as np
import time
# more connectstr
# "dbname='sqlbook' user='postgres' host='localhost' password='xxxx'"
```

```python
def get_connect_str(filename):
    inputfile = open(filename, 'r')

    for i in inputfile:
        connstr = json.loads(i.strip())
    
    inputfile.close()
    return connstr
```

```python
def db_connect(connstr):
    conn = psycopg2.connect(connstr)    
    conn.set_session(autocommit = True)
    return conn.cursor()
```

```python
localconn = db_connect(get_connect_str('connectstr'))
```

```python
def insert_cat(cursor,index,i):
    sql = """insert into category_names (category_id,category)
          values ({0},'{1}')""".format(index,i)

    cursor.execute(sql)
```

```python
catins = open('catlevels.txt','r')
catlist = []

for c in catins:
    c = c.strip()
    catlist.append(c)
    
catins.close()
```

```python
def cat_mapping(cursor):
    sql = """select * from category_names"""
    cursor.execute(sql)
    catmaps = dict()
    for i in cursor.fetchall():
        catmaps[i[1]]=i[0]
        
    return catmaps
```

```python
catmaps = cat_mapping(localconn)
```

```python
def mapping_categories(cursor, catmaps, line):
    row = [line[0]]
    for i in range(1,len(line)):   
        row.append(str(catmaps[line[i]]))
        
    sql = """insert into categories (nodeid,level_1,level_2,level_3,level_4,level_5)
             values ({0})""".format(",".join(row))
    
    cursor.execute(sql)
```

```python
for i in range(1,len(catlist)):
    row = catlist[i].split('\t')
    mapping_categories(localconn,catmaps,row)
```

```python
mapping_categories(localconn,catmaps)
```

```python
def extract_categories(cursor):
    outputs = open('catmapping.txt','w')
    sql = "select * from categories"
    cursor.execute(sql)
    
    for i in cursor.fetchall():
        outputs.write("{0}\n".format("\t".join([str(x) for x in i])))
        
    outputs.close()
```

```python
extract_categories(localconn)
```

```python
def get_category_mapping(cursor):
    sql = """select * from category_names order by category_id"""
    cursor.execute(sql)
    
    with open("numcats.txt", 'w') as outcatmap:
        for r in cursor.fetchall():
            cid = str(r[0])
            cat = str(r[1])
            outcatmap.write("{0}\t{1}\n".format(cid,cat))
```

```python
get_category_mapping(localconn)
```
