```python
#select o.zipcode, ol.productid
#from orders o, orderlines ol
#where o.orderid = ol.orderid

df = pd.read_sql("select o.zipcode, ol.productid, ol.orderlineid  from orders o, orderlines ol  where o.orderid = ol.orderid", con=conn)
df.head()
```

```python
zipcode = np.array(df['zipcode'])
productid = np.array(df['productid'])

matrix = pd.crosstab(zipcode, [productid])
matrix
```

```python
zips = np.array(list(matrix.T))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=100)
data_transformed = pca.fit_transform(matrix[matrix.columns.difference(['row_0'])] )
data_transformed = pd.DataFrame(data_transformed)
```

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(data_transformed)
clusters = kmeans.labels_

clusters_df = pd.DataFrame(clusters)
```

```python
clusters_df['labels'] = zips

test = ['10001', '10002']

print ('Example clusters:')
for word in test:
    clust_num = clusters_df[clusters_df['labels'] == word].values[0][0] 
    print ('\n'+str(word)+':')
    print (clusters_df[clusters_df[0] == clust_num]['labels'].values)
```
