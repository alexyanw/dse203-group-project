```python
from dora.api import DataExplorer

import pandas as pd
import matplotlib.pyplot as plt
from datetime import date, timedelta

explorer = DataExplorer()
```

## Basic Order Stats Aggregated by Product

```python
response = explorer.orders.statsByProduct(order_by='numunits_sum')

stats_df = pd.DataFrame(
    data=response.results, 
    columns=response.columns)

stats_df.head()
```

```python
stats_df[:100].plot.bar(x='asin',y='numunits_sum', xticks=[], figsize=(12,7))
plt.show()
```

## Product Rating Distributions

```python
ratings = explorer.products.ratingsDistribution()
ratings_df = pd.DataFrame(
    data=ratings.results, 
    columns=ratings.columns)
ratings_df.head()
```

```python
ratings_df[:10].plot.bar(
    x='asin',
    y=[
        'one_star_votes',
        'two_star_votes',
        'three_star_votes',
        'four_star_votes',
        'five_star_votes'
    ], figsize=(12,7))

plt.show()
```

### Ratings Distributions Over Time

```python
start = date(2006,5,1)
delta = timedelta(weeks=4)
dates = [start + (delta*x) for x in range(0, 36)]

fig, ax = plt.subplots(12, 3, 
    sharex=True, 
    sharey=True, 
    figsize=(20,24))

row = -1

for i in range(len(dates)-1):
    if i%3 == 0:
        row += 1

    ratings = explorer.products.ratingsDistribution(
        min_date=dates[i],
        max_date=dates[(i+1)], 
        asin=['0002007770'])
    
    if len(ratings.results) == 0:
        continue     
    
    ratings_df = pd.DataFrame(
        data=ratings.results, 
        columns=ratings.columns)

    
    ratings_df.plot.bar(
        ax=ax[row][i%3],
        x='asin',
        y=[
            'one_star_votes',
            'two_star_votes',
            'three_star_votes',
            'four_star_votes',
            'five_star_votes'
        ])
    ax[row][i%3].legend().remove()
    ax[row][i%3].set_title(dates[i])
    

plt.show()
```
